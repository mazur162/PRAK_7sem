```r
install.packages("plotrix")
library(plotrix)

ARMARoots <-function (ar,col)
{
  p <- c(1,-ar)
  r=polyroot(p) # solve the polynomial equation
  rRe<-Re(r)
  rIm<-Im(r)
  plot(rRe, rIm, asp=1, lwd = 8, xlab="Re", ylab="Im", main = "Circle |z|=1 and Roots", pch = 20, col = col, xlim=c(min(rRe,-1), max(rRe,1)), ylim=c(min(rIm,-1),max(rIm,1)))
  draw.circle(0,0,radius=1,border="black",col = "grey")
  points(rRe,rIm,lwd=8,col= col)
  abline(v=0)
  abline(h=0)
}

phi1<-0.6
phi2<--0.2
phi3 <- 0.6
par(mfrow = c(1, 1), cex = 0.9)
ARMARoots(c(phi1,phi2,phi3),"green")
```
