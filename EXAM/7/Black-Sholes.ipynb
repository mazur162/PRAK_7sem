## Модель Блека-Шольца и вычисление дельта портфеля

Инвестор продал 10 колл и 10 пут опционов на акцию компании XXX цена которой в данный момент 1000 рублей, страйк 1000 рублей (такой портфель называется страдл) срок до экспирации полгода, волатильность в данный момент 40%, ставка безрисковая 15%, дивидендов нет. Найти оптимальную цену (цена по модели Black-Sholes) полученную за данный портфель. Сколько акций компании XXX надо купить или продать, чтобы застраховать данный портфель опционов от изменения цен. (т.е. найти δ
 нейтральный портфель акций и опционов)

```r
# не работает в гугл колаб((, но заработает, если сделать install.packages("fBasics")
install.packages("fOptions", repos="http://R-Forge.R-project.org")
library(fOptions)
```

```r

S <- 1000 # цена актива
T <- 0.5 # время в годах
sigma <- 0.4 # волатильность
r <- 0.15 # безрисковая годовая ставка
b <- 0 # the annualized cost-of-carry rate, комиссии
X <- 1000 # цена исполнения - страйк
TypeFlag <- "c"
call <- GBSOption(TypeFlag, S = S, X = X, Time = T, r = r, b = b, sigma = sigma)
summary(call)


TypeFlag <- "p"
put <- GBSOption(TypeFlag, S = S, X = X, Time = T, r = r, b = b, sigma = sigma)
summary(put)


# стоимость портфеля в нашей задаче
10 * call@price + 10 * put@price
```

```r
# вычисляем delta опционов

TypeFlag = 'c'
callDelta <- GBSGreeks(Selection = 'delta',TypeFlag=TypeFlag, S = S, X = X, Time = T, r = r, b = b, sigma = sigma)
callDelta


TypeFlag = 'p'
putDelta <- GBSGreeks(Selection = 'delta',TypeFlag, S = S, X = X, Time = T, r = r, b = b, sigma = sigma)
putDelta


# совокупный delta портфеля:
count <- 10 * callDelta + 10 * putDelta

# значит надо купить -count акций
```
