```r
library(fOptions)
```

```r
# количество опционов в портфеле
num_call <- 10
num_put <- 10
```

```r
S <- 1000 # цена акции
T <- 0.5 # вермя экспирации в годах
sigma <- 0.4 # волатильность
r <- 0.15 # безрисковая ставка
b <- 0 # транзакционные расходы (их нет в Б-Ш, если есть, то в % годовых)
X <- 1000 # страйк

call <- GBSOption("c", S = S, X = X, Time = T, r = r, b = b, sigma = sigma)
put <- GBSOption("p", S = S, X = X, Time = T, r = r, b = b, sigma = sigma)
```

Формула Блэка-Шольца:
$S \cdot N(d) - X\exp{(-r(T - t))}N(d - \sigma\sqrt{T - t}) $, \
где $\displaystyle d = \left[\ln{\cfrac{S}{X}} + (r + \cfrac{\sigma^2}{2})(T−t)\right]\cfrac{1}{\sigma \sqrt{T − t}}$

```r
summary(call)
```

```r
summary(put)
```

```r
num_put * put@price + num_call * call@price
```

Вега портфеля $-$ это скорость изменения величины портфеля относительно изменения волатильности.\
$Vega = \cfrac{\partial \Pi}{\partial \sigma} $

```r
call_vega <- GBSGreeks(Selection = "vega", TypeFlag = "c", S = S, X = X, Time = T, r = r, b = b, sigma = sigma)
put_vega <- GBSGreeks(Selection = "vega", TypeFlag = "p", S = S, X = X, Time = T, r = r, b = b, sigma = sigma)
```

Вега нашего портфеля

```r
num_call * call_vega + num_put * put_vega
```
