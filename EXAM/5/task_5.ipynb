```r
# Матрица ковариации шума

omega <- c(1,0,0)
omega <- rbind(omega,c(0,2,0))
omega <- rbind(omega,c(0,0,4))
omega
```

```r
Phi1 <- c(-0.1, 0.1, 0)
Phi1 <- rbind(Phi1,c( 0.1, 0.1, 0.3))
Phi1 <- rbind(Phi1,c( 0.1, 0.3, 0.2))
Phi1
```

```r
Phi2 <- c(0.2, 0.3, 0.2)
Phi2 <- rbind(Phi2,c(0, 0.3, 0.2))
Phi2 <- rbind(Phi2,c(0.6,-0.4, 0.6))
Phi2
```

```r
Phi <- cbind(Phi1,Phi2)
Phi
```

```r
# Для проверки стационарности формируем матрицу F

I <- diag(1,3,3)
Zero <- diag(0,3,3)
IZ <- cbind(I,Zero)
F <- rbind(Phi,IZ)
F
```

```r
# Находим модули собственных значений матрицы F

eigen_val <- eigen(F,only.values= TRUE)
modev <- Mod(eigen_val$values)
barplot(modev,col = "blue",main = " Abs value of eigen values")
abline(h=1,col = "black")
```

Процесс VAR(p) является стационарным тогда и только тогда, когда все собственные значения матрицы F по модулю меньше единицы |λ|<1. Одно из собственных значений > 1 => Вывод - модель не стационарна
